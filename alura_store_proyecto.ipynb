# 1. Análisis de facturación

```python
ingresoTienda = tienda["Precio"].sum()
print(f"El ingreso total de la tienda es: {ingresoTienda:,.2f}")

ingresoTienda2 = tienda2["Precio"].sum()
print(f"El ingreso total de la tienda 2 es: {ingresoTienda2:,.2f}")

ingresoTienda3 = tienda3["Precio"].sum()
print(f"El ingreso total de la tienda 3 es: {ingresoTienda3:,.2f}")

ingresoTienda4 = tienda4["Precio"].sum()
print(f"El ingreso total de la tienda 4 es: ${ingresoTienda4:,.2f}")


```

```python
ingresos = {
    "Tienda 1": ingresoTienda,
    "Tienda 2": ingresoTienda2,
    "Tienda 3": ingresoTienda3,
    "Tienda 4": ingresoTienda4
}

print(ingresos)
```

```python
plt.figure()
plt.bar(ingresos.keys(), ingresos.values())

plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingresos")

plt.show()
```

    Analisis de Ingresos
En el grafico se aprecia, que la tienda 1 es la que representa mayores ingresos, y la tienda 4 la que tiene menores ingresos, lo que nos dice que esta presentado un menor desempeño

# 2. Ventas por categoría

```python
t1 = tienda.copy()
t2 = tienda2.copy()
t3 = tienda3.copy()
t4 = tienda4.copy()


t1["Tienda"] = "Tienda 1"
t2["Tienda"] = "Tienda 2"
t3["Tienda"] = "Tienda 3"
t4["Tienda"] = "Tienda 4"

# Unir todas las tiendas en un solo DataFrame
todas_las_tiendas = pd.concat([t1, t2, t3, t4], ignore_index=True)

#  Agrupar por Categoría y Tienda
comparacion = (
    todas_las_tiendas.groupby(["Categoría del Producto", "Tienda"]).size().unstack().fillna(0)
)

# Agregar total global por categoría
comparacion["Total Global"] = comparacion.sum(axis=1)

# Ordenar por las categorías más vendidas globalmente
comparacion = comparacion.sort_values("Total Global", ascending=False)

# Graficar (sin la columna Total Global)
comparacion.drop(columns="Total Global").plot(kind="bar", figsize=(10,6))

plt.title("Ventas por Categoría en Cada Tienda")
plt.ylabel("Cantidad de Productos Vendidos")
plt.xlabel("Categoría")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

# 3. Calificación promedio de la tienda

```python
# Como ya tenia todas las tiendas concatenadas

calificacion_promedio= (
    todas_las_tiendas.groupby("Tienda")["Calificación"].mean()
)

print(calificacion_promedio)

ax = calificacion_promedio.plot(kind="bar")

plt.title("Calificación Promedio por Tienda")
plt.ylabel("Promedio de Calificación")
plt.xlabel("Tiendas")
plt.ylim(3,5)

for i, valor in enumerate(calificacion_promedio):
    ax.text(
        i,                 # posición horizontal
        valor + 0.05,       # posición vertical (un poco arriba)
        f"{valor:.2f}",     # formato con 2 decimales
        ha="center"         # alineación horizontal
    )
plt.show()
```
